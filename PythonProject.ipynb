```python
#1.How many are there in each Team and the percentage splitting with respect to the total employees.
team_employees=ABC_company["Team"].value_counts()
team_employees
```

```python
percentage=(team_employees/ABC_company.shape[0])*100
percentage
```

```python
#2.Segregate the employees w.r.t different positions.
ABC_company["Position"].value_counts()
```

```python
ABC_company_position=ABC_company.groupby('Position')
for Position,group in ABC_company_position:
   print(f'Employees in {Position}:')
   print(group)
```

```python
#3.Find from which age group most of the employees belong to
age_group=ABC_company['Age'].value_counts().head()
age_group
```

```python
#4.Find out under which team and position, spending in terms of salary is high
grouped=ABC_company.groupby(['Team','Position'])['Salary'].sum()
grouped
highest_spending=grouped.idxmax()
highest_spending
```

```python
#5.Find if there is any correlation between age and salary , represent it visually.
correlation= ABC_company[['Age', 'Salary']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Age and Salary')
plt.show()
```
